```python
def show_img(img):
    plt.figure(figsize=(10,10))
    plt.imshow(img, cmap="gray")
    plt.show()
```

```python
matrix = np.array([
    [1,0,-1],
    [0,0,0],
    [-1,0,1]
])
matrix2 = 1/9*np.ones(shape=(3,3))
# filtered = filters.median_filter(img, size=5)
filtered = filters.convolve(img, matrix2)
filtered = filters.convolve(filtered, matrix)
```

```python
show_img(filtered)
```
